```python
dataset = FaceDataset("/mnt/bn/data-tns-live-llm/leon/experiments/llm/face/cropped_second_stage_imgs/")
print(len(dataset))
```

```python
import cv2 as cv
from torchvision.transforms import transforms
img = cv.imread("/opt/tiger/SimCLR/label.jpg")
print(img.shape)
normalize = transforms.Normalize(
    mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
)
resize = transforms.Resize((224,224))
transform = transforms.Compose([
    transforms.ToTensor(),
    resize, 
    normalize
    ])
```

```python
img = transform(img)
print(img.shape)
print(img)
print(torch.mean(img))
```

```python
from torch.utils.data import DataLoader
train_loader = DataLoader(dataset, batch_size=16, shuffle=True,
num_workers=10, pin_memory=True, drop_last=True)
```

```python
print(len(train_loader))
```
